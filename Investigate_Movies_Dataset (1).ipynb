### Research Question 1 (Which year has the highest release of movie!)

```python
# Group the data by release year and check the last five rows
year_count = df.groupby('release_year').count()['id']
print(year_count.tail())

# Make group of the data according to their release year and count the total number of movies in each year and pLot.
df.groupby('release_year').count()['id'].plot(xticks = np.arange(1960,2016,5))

#set the figure size and labels
sns.set(rc={'figure.figsize':(10,5)})
plt.title("Year Vs Number Of Movies", fontsize = 14)
plt.xlabel('Release year', fontsize = 12)
plt.ylabel('Number Of Movies', fontsize = 12)

#set the style sheet
sns.set_style("darkgrid")
```

### Answer: The line chart shows that year 2014 has the highest release of movies (700) followed by year 2013 (659) and year 2015 (629)

### Research Question 2  (Movie with Highest And Lowest Profit?)

```python
# Check for the Movie with highest and lowest profit

df['Profit'] = df['revenue'] - df['budget']
```

```python
def find_minmax(x):
    #use the function 'idmin' to find the index of lowest profit movie.
    min_index = df[x].idxmin()
    #use the function 'idmax' to find the index of Highest profit movie.
    high_index = df[x].idxmax()
    high = pd.DataFrame(df.loc[high_index,:])
    low = pd.DataFrame(df.loc[min_index,:])
    
    #print the movie with high and low profit
    print("Movie Which Has Highest "+ x + " : ",df['original_title'][high_index])
    print("Movie Which Has Lowest "+ x + "  : ",df['original_title'][min_index])
    return pd.concat([high,low],axis = 1)

#call the find_minmax function.
find_minmax('Profit')
```

### Answer: The analysis shows that Avatar had the highest profit while the Warrior's Way had the lowest profit

### Research Question 3  (Top 10 movies which earn highest profit?)

```python
# To check the top 10 movies that earn the highest profit.

top_movies = pd.DataFrame(df['Profit'].sort_values(ascending = False))
top_movies['original_title'] = df['original_title']
data = list(map(str,(top_movies['original_title'])))
x = list(data[:10])
y = list(top_movies['Profit'][:10])

#make a plot using pointplot for top 10 profitable movies.
ax = sns.barplot(x=y,y=x)

#setup the figure size
sns.set(rc={'figure.figsize':(10,5)})
#setup the title and labels of the plot.
ax.set_title("Top 10 Profitable Movies",fontsize = 15)
ax.set_xlabel("Profit",fontsize = 13)
sns.set_style("darkgrid")
```

### Answer: The top 10 profitable movies are: Avatar, Star Wars: The Force Awakens, Titanic, Jurassic World, Furious 7, The Avengers, Harry Potter and the Deathly Hallows Part 2, Avenger:Age of Ultron, Frozen, The Net.

### Research Question 10 (Which Month Released Highest Number Of Movies In All Of The Years?)

```python
#Extract the months from the years 
month_release = df['release_date'].dt.month

#count the movies in each month using value_counts().
number_of_release = month_release.value_counts().sort_index()
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
number_of_release = pd.DataFrame(number_of_release)
number_of_release['month'] = months

#change the column name of the new dataframe 'number_of_release'
number_of_release.rename(columns = {'release_date':'number_of_release'},inplace=True)

#plot the bar graph using plot.
number_of_release.plot(x='month',kind='bar',fontsize = 11,figsize=(10,6))

#set the labels and titles of the plot.
plt.title('Months vs Number Of Movie Releases',fontsize = 15)
plt.xlabel('Month',fontsize = 13)
plt.ylabel('Number of movie releases',fontsize = 13)
sns.set_style("darkgrid")

```

### Answer: The chart shows that the month of September and October are the month with highest number of movies released in all the years  

### Research Question 11 (Which Month made the Highest Average Revenue In All Of The Years?)

```python
#Extract the date from release date
month_release = pd.DataFrame(month_release)

#change the column name of the new dataframe 'month_release'.
month_release.rename(columns = {'release_date':'release_month'},inplace=True)

#add a new column 'revenue' in the dataframe 'month_release'.
month_release['revenue'] = df['revenue']

#make the group of the data according to their month and calculate the mean revenue of each month.
mean_revenue = month_release.groupby('release_month').mean()
mean_revenue['month'] = months

#make the bar plot using pandas plot function.
mean_revenue.plot(x='month',kind='bar',figsize = (10,6),fontsize=11, color='gold')

#setup the title and lables of the plot.
plt.title('Average revenue by month (1960 - 2015)',fontsize = 15)
plt.xlabel('Month',fontsize = 13)
plt.ylabel('Average Revenue',fontsize = 13)
sns.set_style("darkgrid")
```

### Answer: The chart shows that the month of June and May are the months with Highest Average Reveune in all the years

```python
def plot_correlation_map(df):
    corr = df.corr()
    _ , ax = plt.subplots(figsize =(12,10))
    cmap = sns.diverging_palette( 240 , 10 , as_cmap = True )
    _ = sns.heatmap(corr,cmap = cmap, square=True, cbar_kws={ 'shrink' : .9 }, ax=ax, annot = True, annot_kws = {'fontsize':12})
```

```python
plot_correlation_map(df[['popularity','budget','revenue','runtime','vote_count','vote_average']])
```
